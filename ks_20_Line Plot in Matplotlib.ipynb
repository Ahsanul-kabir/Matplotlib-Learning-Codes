```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0,2.0,0.01) # X-axis

s = 1 + np.cos(2*np.pi*t)   # Y-axis ----->> {1 + cos(2*pi*t)}

plt.plot(t,s,'--')          # (--/*/green/blue) --->> jeita use kormu oita asbe

plt.grid()
plt.xlabel('Time(t)')
plt.ylabel('Voltage(mV)')
plt.title('Cosine wave plot(Cos(x))')

plt.show()
```
