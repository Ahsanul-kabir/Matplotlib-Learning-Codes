```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0,5.0)  #time # Linear Space
x2 = np.linspace(0.0,2.0)  ## Linear Space


y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1) # (row =2 ,col =1,First_Position of Grap)    # Row & Column format e kora lage
plt.plot(x1,y1,'o-')
plt.title('Subplot-1')
plt.xlabel('x1')
plt.ylabel('Amp((y1))')


plt.subplot(2,1,2) # (row =2 ,col =1, 2nd_Position of Grap -->> First Position ta nise holo == x1,y1 e...tai x2,y2 2nd Position nibe)
plt.plot(x2,y2,'.-')
plt.title('Subplot-2')
plt.xlabel('x2')
plt.ylabel('Amp((y2))')

plt.show()
```
