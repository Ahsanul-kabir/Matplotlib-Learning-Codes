```python
import matplotlib.pyplot as plt

activities= ['eat','sleep','work','play']

slices = [3,7,8,9]

colors_list = ['r','g','m','b']       # 'startangle' mane 90 degree angle e start hbe chart # 'shadow' just like 3D view er jonno 
                                     # 'explode' diya oi "Red" part ta bahit hoise oitar kaj kora hoy # 'autopct' mane Parcent e show kora
plt.pie(slices, labels=activities, colors = colors_list, startangle =90, shadow =True,explode=(0.2,0,0,0), autopct = '%1.2f%%')

plt.legend()

plt.show()
```
