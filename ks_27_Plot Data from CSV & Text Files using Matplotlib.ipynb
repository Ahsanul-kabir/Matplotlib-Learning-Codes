```python
import matplotlib.pyplot as plt

import csv

x= []
y= []

with open('text.txt','r') as textfile:   # for text Files
    plots = csv.reader(textfile)
    for col in plots:
        x.append(col[0])   # Value column Format e neya hoise
        y.append(col[1])
 


plt.plot(x,y,label='Files',color='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Test Graph')


plt.legend()

plt.show()

```

```python
import matplotlib.pyplot as plt

import csv

x= []
y= []

with open('text.csv','r') as csvfile:   # For csv Files
    plots = csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])
 


plt.plot(x,y,label='Files',color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Test Graph')


plt.legend()

plt.show()
```
