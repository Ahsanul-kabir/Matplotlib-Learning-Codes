```python
import matplotlib.pyplot as plt

import numpy as np

x = np.arange(0,2*np.pi,0.1)  # like circle

y1 = np.sin(x)

y2 = np.cos(x)

plt.plot(x,y1,label='Sin')

plt.plot(x,y2,label='Cos')


plt.xlabel('x-axis')

plt.ylabel('y-axis')

plt.title('Sine & Cosine Functions')


plt.legend()
plt.show()
```
