```python
import matplotlib.pyplot as plt

ages = [2,50,40,70,30,30,51,52,30,44,60,80,90,3,60,66,8,22,23,44,55,83]

rang = (0,100) # range

bins = 10  # Height 10 bar hbe...tar kom o thake pare "Value kom hole"

plt.hist(ages,bins,rang,color='m',histtype='bar' ,rwidth = 0.7)

plt.xlabel('x-axis')

plt.ylabel('y-axis')

plt.title('Histogram Plot')

plt.show()
```
